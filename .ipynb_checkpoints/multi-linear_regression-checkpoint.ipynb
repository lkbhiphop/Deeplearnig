```python
import tensorflow as tf
import numpy as np
```

```python
x1_data = [10., 12., 15., 16., 78.]
x2_data = [13., 22., 15., 17., 78.]
x3_data = [85., 184., 19., 12., 78.]
y_data = [123.,345.,234.,456.,789]

# 위의 데이터를 받을 그릇
x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

# weight를 담을 그릇(variable)
w1 = tf.Variable(tf.random_normal([1]),name = "weight1")
w2 = tf.Variable(tf.random_normal([1]),name = "weight2")
w3 = tf.Variable(tf.random_normal([1]),name = "weight3")
b = tf.Variable(tf.random_normal([1],name = 'bias'))
# 초기화
hypothesis = x1 * w1 + x2 * w2 + x3 * w3 + b
cost = tf.reduce_mean(tf.square(hypothesis-Y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 1e-5)
train = optimizer.minimize(cost)
```

```python
# 2만번 돌렸을 때 가장 최적화된 hypothesis를 보여준다. 
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(20000):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train], feed_dict = {x1:x1_data,x2:x2_data,x3:x3_data,Y:y_data})
    
    
    if step % 100 == 0:
        print( "Step: ", step, "Cost : ", cost_val, "Hy :" , hy_val)
```

```python
#learning 100이면 overshooting이나고
# 1e-30 하면 30000에서 내려가지 않는다. 
# 보통 0.1 ~ 0.00001을 넣는다.
```
