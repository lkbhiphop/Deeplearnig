```python
import tensorflow as tf
import numpy as np
```

```python
xy = np.loadtxt("diabetes.csv", delimiter=',',dtype = np.float32)
```

```python
xy.shape
```

```python
x_train = xy[:559,0:-1] # 559 * 8
y_train = xy[:559,-1:]  # 559 * 1
x_test = xy[559:,0:-1]
y_test = xy[559:,-1:]
```

```python
x_test.shape
```

```python
y_test
```

```python
# 2. 데이터 그릇만들기
X = tf.placeholder(tf.float32, shape = [None,8])
Y = tf.placeholder(tf.float32, shape = [None,1])
```

```python
# 3. w랑 b 만들기
W = tf.Variable(tf.random_normal([8,1]),name = "Weight")
b = tf.Variable(tf.random_normal([1]),name = "bias")
# 4. 식 만들기 hypothesis -- y값을 뽑아낼 식 만들기 
hypothesis = tf.sigmoid(tf.matmul(X,W)+b)

# 5. cost(기준 만들기)
cost = -tf.reduce_sum(tf.reduce_mean(Y * tf.log(hypothesis) + (1-Y) * tf.log(1-hypothesis) ))
# 6. optimizer 만들기 
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
# 7. train 만들기
train = optimizer.minimize(cost)
# 8. 분류 모델 추가하기
predicted = tf.cast(hypothesis > 0.5, dtype = tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype = tf.float32))
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
for step in range(2000):
    c_v, hy_v, _ = sess.run([cost, hypothesis, train], feed_dict= {X:x_train,Y:y_train})
    if step % 100 == 0 :
        print("Step: ",step, "Cost:",c_v,"hy_v:", hy_v)
```

```python
sess.run(predicted, feed_dict = {X:x_test})
```

```python
print(sess.run(accuracy,feed_dict = {X:x_test, Y:y_test}))
```
