```python
# np.arrange(숫자) --> 0부터 숫자가지 배열을 만든다.
# np.arrange(숫자).reshape(4,3)
arr = np.arange(12).reshape(4,3)
print(arr)
row_mean = arr.mean(axis=1).reshape(4,1)
# 빼기하면 브로드캐스팅된다!!!!
k = arr-row_mean
# normalize하는데 쓰이기도한다.
print(k)
# numpy/pandas/matplotlib
```
